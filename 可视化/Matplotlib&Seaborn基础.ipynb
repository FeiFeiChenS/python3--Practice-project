```python
#导入库
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
##散点图
x=np.random.randn(500)
y=np.random.randn(500)
#matplotlib
plt.scatter(x,y)
plt.show()
#seaborn
# sns.scatterplot(x,y)        #下面是seaborn的图，会比上面好看点，还可以显示数据分布
df=pd.DataFrame(np.random.randn(200,2),columns=["x-axis","y-axis"])    #好像不能设置subplot图像位置
sns.jointplot(x=df['x-axis'],y=df['y-axis'],                #联合图,含x，y的概率分布
              data=df,
              color='r',
              kind="scatter")                #kind可以改成hex-蜂窝，kde-核密度
# plt.subplot(211)
# sns.scatterplot(data=data)      #这个图像错了，应该是吧x、y当做不同图例，把值和下标当做一组数据
sns.jointplot(x=df['x-axis'],y=df['y-axis'],
              data=df,
              color='b',
              kind="hex") 
sns.jointplot(x=df['x-axis'],y=df['y-axis'],
              data=df,
              color='g',
              kind="kde") 
```

```python
##折线图
from matplotlib.font_manager import FontProperties #字体管理器
myfont=FontProperties(fname="msyh.ttc")
y=np.random.randint(0,100,100)
x=[_ for _ in range(len(y))]
#matplotlib
plt.plot(x,y)
plt.xlabel("x轴",fontproperties=myfont)
plt.ylabel("y轴",fontproperties=myfont)
plt.title("随机数",fontproperties=myfont)
plt.show()
#seaborn
data={"x-axis":x,"y-axis":y}
df=pd.DataFrame(data)
sns.lineplot(x=x,y=y,data=df)
plt.xlabel("x轴",fontproperties=myfont)
plt.ylabel("y轴",fontproperties=myfont)
plt.title("随机2",fontproperties=myfont)
```

```python
#相关系数+热力图
data=np.random.randn(8,7)
df=pd.DataFrame(data.round(2),columns=["A","B","C","D","E","F","G"])
persion=df.corr().round(2)
sns.heatmap(persion,annot=True)    #annot绘制数值
```

```python
#feature importance
data=np.random.randn(8,7)
df=pd.DataFrame(data.round(2),columns=["A","B","C","D","E","F","G"])
persion=df.corr().round(2)
sns.heatmap(persion,annot=True)    #annot绘制数值
```
