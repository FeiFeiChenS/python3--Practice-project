```python
# -*- coding:gbk -*-
#加载数据与模型库
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/机器学习/K-MEANS_球队聚类/team_cluster_data.csv",encoding="gbk")  #编码问题,设置encoding参数
train=df.iloc[:,1:]
print(df)
```

```python
#sklearn三部曲
km=KMeans(n_clusters=3)  #簇的数量，或者说聚类数，属于超参数，自己设置
km.fit(train)
km_class=km.predict(train)
print(km_class)
```

```python
##分类数据的可视化，三维好像比较难搞，试试两两可视化
import seaborn as sns
```

```python
#colab设置中文字体
from matplotlib.font_manager import FontProperties
my_font=FontProperties(fname="/content/drive/MyDrive/机器学习/K-MEANS_球队聚类/MSYH.TTC")
```

```python
#在散点图上批量添加注释
import matplotlib.pyplot as plt
# print(my_font.get_name()) ##Microsoft YaHei
# sns.set(font=my_font.get_name())
# sns.set_style("whitegrid",{"font.sans-serif":my_font.get_name()})
sns.set_theme(style="whitegrid",font=my_font.get_name())
km_scatter=sns.scatterplot(x=df["2019国际排名"],y=df["2018世界杯排名"],hue=km_class,style=km_class)   #hue设置分类，style设置不同标记点
for i in range(len(df["国家"])):
  km_scatter.text(df["2019国际排名"][i]+0.2,df["2018世界杯排名"][i],df["国家"][i])
# plt.xticks(fontproperties=my_font)
# plt.xlabel(x=df["2019国际排名"],fontproperties=my_font)
# plt.ylabel(fontproperties=my_font)
plt.show()
```
