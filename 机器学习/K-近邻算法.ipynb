```python
# 不使用机器学习.库，实现K-近邻算法

import numpy as np
from matplotlib import pyplot as plt
x_trains=np.array([[1,101],[5,89],[108,5],[115,8],[100,10]])
y_lables=['爱情片','爱情片','动作片','动作片',"动作片"]

testdata=np.array([[80,80]])
k=3
x=[x for [x,y] in x_trains]
y=[y for [x,y] in x_trains]
plt.scatter(x,y)
plt.show()
def classifyKnn(testdata,x_trains,y_lables,k):
    testlable=[]
    for i in testdata:
        dist=((x_trains-i)**2).sum(axis=1)
        dindex=dist.argsort()
        classlable=[]
        for ind in range(k):
            index=dindex[ind]
            classlable.append(y_lables[index])
        for item in range(len(classlable)):
            count=classlable.count(classlable[item])
            if count/(len(classlable))>=0.5:
                testlable.append(classlable[item])
                break
    return testlable

if __name__=="__main__":
    testlable=classifyKnn(testdata,x_trains,y_lables,3)         # K值为奇数，且数字不能过大（偶数会存在不能分类【50%】的问题）
    print(testlable)
        
        
```

```python
#用scikit learn机器学习库，实现K-近邻算法
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

x_trains=[[1,101],[5,89],[108,5],[115,8],[100,10]]
y_lables=['爱情片','爱情片','动作片','动作片',"动作片"]
testdata=[[3,39],[88,36],[9,99]]

myKNN=KNeighborsClassifier(n_neighbors=3)
myKNN.fit(x_trains,y_lables)
myKNN.predict(testdata)
```

```python
#用sklearn 鸢尾花数据集，实现knn分类算法

from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

iris=datasets.load_iris()
x_data=iris["data"]
y_lable=iris["target"]

x_trains,x_test,y_trains,y_test=train_test_split(x_data,y_lable,test_size=0.3)


#构建分类器
myknn = KNeighborsClassifier()
myknn.fit(x_trains,y_trains)
y_pre = myknn.predict(x_test)

print(y_pre)
print(y_test)

diff=0
for i in range(len(y_pre)):
    if y_pre[i] !=y_test[i]:
        diff+=1
print("/d%" %diff/len(y_pre)*100)
```
