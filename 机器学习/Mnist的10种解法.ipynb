```python
#加载数据
from sklearn import datasets  #好吧，sklearn的dataseets里面不包含minist
```

```python
import pandas as pd
df_train=pd.read_csv("/content/sample_data/mnist_train_small.csv")
df_test=pd.read_csv("/content/sample_data/mnist_test.csv")
```

```python
df_test.head(5)
```

```python
y_train=df_train["6"]
y_test=df_test["7"]
```

```python
x_train=df_train.drop("6",axis=1)
x_test=df_test.drop("7",axis=1)
```

```python
#1.用knn算法
from sklearn.neighbors import KNeighborsClassifier
#初始化模型
model=KNeighborsClassifier()
#train fit
model.fit(x_train,y_train)
#测试数据
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)
```

```python
#knn算法的缺点：速度太慢了，黄花菜都凉了
#改进点：（1）初始化模型的参数可以调整；（2）数值没有归一化处理
#由于速度太慢了，改进方法暂不测试
```

```python
#2.决策树
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)
```

```python
#决策树的缺点：正确率较低，且预测结果比较依赖于数据集的拆分
```

```python
#3.LR逻辑斯蒂回归,注意从线性模型中导入,模型的首字母都大写
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)
```

```python
#逻辑斯蒂回归中遇到的问题：迭代次数受到限制，模型没有收敛
```

```python
#4.1 朴素贝叶斯
from sklearn.naive_bayes import GaussianNB   #假设数据服从高斯分布，高斯分布又名正态分布
model=GaussianNB()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)      #这个正确率太差了
```

```python
#4.2 朴素贝叶斯
from sklearn.naive_bayes import MultinomialNB #多项式分布的效果比正态分布的效果好
model=MultinomialNB()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)      
```

```python
#5.1 SVM支持向量机，线性分类器
from sklearn.svm import LinearSVC
model=LinearSVC()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)     #速度较慢，同样需要归一化处理
```

```python
#5.2 SVM支持向量机，非线性分类器
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)     #效果比线性分类器要好，但是速度同样比较慢
```

```python
#6.adaboost
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)     #效果不理想，是过拟合问题，还是参数问题
```

```python
#7.随机森林
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
true_rate=(y_predict==y_test).sum()/len(y_predict)
print(true_rate)    #效果好，速度快,bagging算法，多个分类器平行
print(model.score(x_test,y_test))
```
