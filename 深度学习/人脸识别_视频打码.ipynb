####cv2.VideoCapture.get()的参数

```python
#获取视频的参数，cv2.VideoCapture.get()  
# 0-视频文件的当前位置（播放）以毫秒为单位;
# 1-基于以0开始的被捕获或解码的帧索引;
# 2-视频文件的相对位置（播放）：0=电影开始，1=影片的结尾。;
# 3-在视频流的帧的宽度;
# 4-在视频流的帧的高度;
# 5-帧速率;
# 6-编解码的4字-字符代码;
# 7-视频文件中的帧数;
# 8-返回对象的格式;
# 9-返回后端特定的值，该值指示当前捕获模式;
# 10-图像的亮度(仅适用于照相机);
# 11-图像的对比度(仅适用于照相机);
# 12-图像的饱和度(仅适用于照相机);
# 13-色调图像(仅适用于照相机);
# 14-图像增益(仅适用于照相机)（Gain在摄影中表示白平衡提升;
# 15-曝光(仅适用于照相机);
# 16-指示是否应将图像转换为RGB布尔标志;
# 17-× 暂时不支持;
# 18-立体摄像机的矫正标注（目前只有DC1394 v.2.x后端支持这个功能;
```

# 图像识别

```python
##自动打码
from matplotlib import pyplot as plt
def addmask(videounsorted_path,videosorted_path,mask_path="/content/drive/MyDrive/深度学习/简单人脸识别：face_recognition/mask.jpg"):
  #读取打码图片
  mask=cv2.imread(mask_path)
  #读取视频
  cap=cv2.VideoCapture(videounsorted_path)    
  v_fps=cap.get(5)    #获取帧速率
  v_hight=cap.get(4)   #获取高度
  v_width=cap.get(3)   #获取宽度

  #设置写视频的参数
  write_size=(int(v_width),int(v_hight))
  write_fourcc=cv2.VideoWriter_fourcc("m","p","4","v")      #VideoWriter_fourcc为视频编解码器
  write_out=cv2.VideoWriter(videosorted_path,write_fourcc,v_fps,write_size)  #四个参数分别是，写入对象(路径)，解码器，帧速率，视频帧大小

  #已知人脸
  know_face=face_recognition.load_image_file(r"/content/drive/MyDrive/深度学习/简单人脸识别：face_recognition/tz.jpg")
  face_feature=face_recognition.face_encodings(know_face)[0]

  #读取并写入视频
  #cv2.VideoCapture().read()从视频或摄像头中读取一帧（即一张图像）,返回是否成功标识ret(True代表成功，False代表失败），img为读取的视频帧
  while(cap.isOpened()):
    #按帧数读取数据
    ret,frame=cap.read()
    if ret: #表示该帧读取成功
      #检测人脸位置
      face_loc=face_recognition.face_locations(frame)    #找到图像上人脸的位置，返回人脸位置(top,right,bottom,left)上右下左顺序的list
      for (top,right,bottom,left) in face_loc:
        # cv2.rectangle(frame,(left,top),(right,bottom),(0,0,255),10)
        # plt.imshow(cv2.cvtColor(frame,cv2.COLOR_RGB2BGR))
        # plt.show()
        unknown_image=frame[top-50:bottom+50,left-50:right+50]
        unknown_feature=face_recognition.face_encodings(unknown_image)[0]
        
        #对比结果
        results=face_recognition.compare_faces([face_feature],unknown_feature)

        #打码
        if results[0] == True:
          mask=cv2.resize(mask,(right-left,bottom-top))   #注意，opencv以左上角为坐标轴原点，往下y轴是正值！！所以这里是bottom-top
          frame[top:bottom,left:right]=mask        #frame索引符合先行后列，行的索引对应是top：bottom，列的索引对应是left：right
      #写入视频
      write_out.write(frame)
    else:
      break

```

```python
# cv2.rectangle参数：#cv2.rectangle(image, start_point, end_point, color, thickness),用于在任何图像上绘制矩形,从facelocation转换（左上，右下）
# image:它是要在其上绘制矩形的图像。
# start_point：它是矩形的起始坐标。坐标表示为两个值的元组，即(X坐标值，Y坐标值)。
# end_point：它是矩形的结束坐标。坐标表示为两个值的元组，即(X坐标值ÿ坐标值)。
# color:它是要绘制的矩形的边界线的颜色。对于BGR，我们通过一个元组。例如：(255，0，0)为蓝色。
# thickness:它是矩形边框线的粗细像素。厚度-1像素将以指定的颜色填充矩形形状。
```
