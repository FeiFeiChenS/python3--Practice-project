```python
import requests
from bs4 import BeautifulSoup
```

```python
#获取url请求信息
request_url=r"http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-1.shtml"
headers={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"}
response=requests.get(request_url,headers=headers)
type(response)
print(response)
```

```python
#解析网页代码
bs=BeautifulSoup(response.text,"html.parser")             #response对象用.text获取内容
print(type(bs))
```

```python
soup=bs.find("table",class_="ar_c ar_c1")              #find_all会作为一个整体输出结果，如果对后面内容还需要细分的，这里调用find函数就可以了
print(type(soup))
print(soup) 
```

```python
datalist=[]
for tr in soup.find_all("tr")[1:]:
  data=[]
  for td in tr: 
    data.append(td.string)
  datalist.append(data)
```

```python
import pandas as pd
pd.DataFrame(datalist,columns=["投诉编号","投诉品牌","投诉车系","投诉车型","问题简述","典型问题","投诉时间","投诉状态"]) 
```
