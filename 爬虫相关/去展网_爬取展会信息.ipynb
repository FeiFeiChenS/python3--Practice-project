```python
##展会名字和链接爬完以后，爬具体展会的内容
df_ex=pd.read_excel(r"C:\Users\fh282\Desktop\exlist.xls",names=["xuhao","name","link"])
df_ex
```

```python
#展会名称、展会时间、展览行业、主办单位、展览面积、展商数量、观众数量、展品范围、展馆名称、展馆地址、联系方式
import re
all_info=[]
for i in range(len(df_ex["name"])):   #range(len(df_ex["name"]))
    one_info=[]  #一条展会的信息
    #1、展会名称已加入
    ex_name=df_ex["name"][i]
    one_info.append(ex_name)
    response=requests.get(df_ex["link"][i],headers=headers)
    bs_info=BeautifulSoup(response.text,"html.parser")
    #2、展会时间
    ex_time=bs_info.find(class_="date").text
    one_info.append(ex_time)
    #3、展览行业
    ex_hy=bs_info.find(class_="con_hy").a.text
    one_info.append(ex_hy)
    #4~7主办单位、展览面积、展商数量、观众数量
    four_to_severn=bs_info.find(class_="zh-description")
    #4.主办单位
    four_to_severn=str(four_to_severn)
    find_unit=re.compile(r"主办单位：</span> (.*)</p>")
    unit=re.findall(find_unit,four_to_severn)      #序号转化字符串
    one_info.append(unit[0])
    #5.展览面积
    find_area=re.compile(r"展览面积：(.*)</span>")
    area=re.findall(find_area,four_to_severn)
    one_info.append(area[0])
    #6、展商数量
    find_merchant=re.compile(r"展商数量：(.*)</span>")
    merchant=re.findall(find_merchant,four_to_severn)
    one_info.append(merchant[0])
    #7、观众数量
    find_audience=re.compile(r"观众数量：(.*)</span>")
    audience=re.findall(find_audience,four_to_severn)
    one_info.append(audience[0])
    #8、展品范围
    content=bs_info.find_all(class_="zh-description")
    scope=""
    for p in content[1]:
        scope+=str(p).replace("<p>","").replace("</p>","").replace("<b>","").replace("</b>","").replace(r"\n","")+"\n"
    one_info.append(scope)
    #9、展馆名称    10、展馆地址    11、联系方式
    place=bs_info.find(class_="zg-msg bd cls")
    if place:
        #9、展馆名称 
        three_info=place.dl
        placename=three_info.dt.text
        one_info.append(placename)
        #10、展馆地址
        find_adress=re.compile(r"展馆地址：</span>(.*)</dd>")
        adress=re.findall(find_adress,str(three_info))
        one_info.append(adress[0])
        #11、联系方式
        find_tel=re.compile(r"联系电话：</span>(.*)</dd>")
        tel=re.findall(find_tel,str(three_info))
        one_info.append(tel[0])
    else:
        one_info.append(" ")
        one_info.append(" ")
        one_info.append(" ")
    all_info.append(one_info)
all_info_pd=pd.DataFrame(all_info,columns=["展会名称","展会时间","展览行业","主办单位","展览面积","展商数量","观众数量","展品范围","展馆名称","展馆地址","联系电话",])
all_info_pd.to_excel(r"C:\Users\fh282\Desktop\allinfo.xls")
```
